## Detecting OR Handling Outliers

### Using Boxplot

```python
titanicDF.isnull().sum()
sb.boxplot(x=titanicDF['Age'])
```

### Using Descriptive Statistics

```python
# Calculate Mean
mean = titanicDF['Age'].mean()
std = titanicDF['Age'].std()
outliers = titanicDF[(titanicDF['Age'] > mean + 3*std) | (titanicDF['Age'] < mean - 3*std)]
print("Mean :  ",mean)
print("Std :  ",std)
print("---------------------------------------------")
print("Outliers :  ",outliers)
```

### Using Interquartile Range (IQR) Method (Most common)

```python
Q1 = titanicDF['Fare'].quantile(0.25)
Q3 = titanicDF['Fare'].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
outliers = titanicDF[(titanicDF['Fare'] < lower_limit) | (titanicDF['Fare'] > upper_limit)]

print("Lower Limit : ",lower_limit)
print("Upper Limit : ",upper_limit)
print("---------------- Outliers --------------------------")
print(outliers)

```
