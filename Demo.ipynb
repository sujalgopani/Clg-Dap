## Histogram
### Histogram Is The Used For the Knowing Frequency Of The Whole Data

### Use Color And bins

```python
mt.hist(studentDF['math score'],bins=10,color='pink',edgecolor='red')
mt.title("Student Univariant Math Score Anlysics")
mt.xlabel("<----------------------Students Number---------------------->")
mt.ylabel("<----------------------Marks Range---------------------->")
mt.show
```

```python

mt.hist(studentDF['reading score'],bins=20,color='pink',edgecolor='red')
mt.title("Student Univariant Reading Score Anlysics")
mt.xlabel("Students Number")
mt.ylabel("Marks Range")
mt.show
```

```python

mt.hist(studentDF['writing score'],bins=20,color='pink',edgecolor='red')
mt.title("Student Univariant Writing Score Anlysics")
mt.xlabel("Students Number")
mt.ylabel("Marks Range")
mt.show()
```

```python

mt.hist(titanicDF['Age'],bins=20,color='lightblue',edgecolor='black')
mt.title("Titanic Univariant Age Anlysics")
mt.xlabel("Age")
mt.ylabel("Frequency")
mt.show()
```

```python
mt.hist(titanicDF['Fare'],bins=10,color='lightblue',edgecolor='black')
mt.title("Titanic Univariant Fare Anlysics")
mt.xlabel("Fare")
mt.ylabel("Frequency")
mt.show()
```

```python
# sb.distplot(titanicDF['Age'],kde='True')
sb.displot(titanicDF['Age']) # without line
mt.show()
sb.displot(titanicDF['Age'],kde='True') # using kde true we can provide lines
mt.show()
```
